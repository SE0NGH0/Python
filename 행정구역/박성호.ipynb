### 1. 202302_202302_주민등록인구및세대현황_월간.csv를 가지고 와서 'data'라는 이름의 데이터 프레임으로 작성하고 5개만 확인합니다.

```python
import pandas as pd
```

```python
data = pd.read_csv('202302_202302_주민등록인구및세대현황_월간.csv', encoding='EUC-KR')
```

```python
data.head()
```

### 2. '행정구역'컬럼을 행정구역명과 행정기관코드로 분리하여 'data' 데이터 프레임에 '행정구역' 컬럼에 행정구역명만 저장하고 5개의 데이터만 확인합니다.

```python
df = pd.DataFrame(data)
```

```python
df['행정구역'] = df['행정구역'].str.replace(r'\(.*\)', '').str.strip()

data.head()
```

### 3. 'data' 데이터 프레임의 컬럼명을 행정구역, 총인구수, 세대수, 세대당인구, 남자, 여자, 남녀비율 로 변경하고 'data'데이터 프레임의 정보를 확인합니다.

```python
data.columns = ['행정구역', '총인구수', '세대수', '세대당인구', '남자', '여자', '남녀비율']
```

```python
data
```

### 4. 남녀비율 컬럼을 제거하고 'data' 데이터프레임 정보를 확인합니다.

```python
data = data.drop(columns=['남녀비율'])
```

```python
data
```

### 5. 행정구역 컬럼의 ‘(’ ‘)’를 제거하고 시,구,동 컬럼을 만들어 분리해 넣고 세종특별자치시 지역의 값만 선택해서 'sejong' 데이터 프레임에 저장합니다.

```python
df = pd.DataFrame(data)
```

```python
split_data = df['행정구역'].str.split(expand=True)
df['시'] = split_data[0]
df['동'] = split_data[1]

sejong = df[df['행정구역'].str.contains('세종특별자치시')]

sejong['시'] = '세종특별자치시'
sejong['동'] = df['동']

sejong
```

### 6. 'sejong' 데이터 프레임에 NAN 값이 있다면 해당 행을 제거하고, 기존 인덱스 값은 제거하고 새로 인덱스를 부여하고 내용을 확인합니다.

```python
sejong = sejong.dropna()

sejong = sejong.reset_index(drop=True)

sejong
```

### 7. 총인구수, 남자, 여자 컬럼의 데이타 타입을 정수형으로 변경하고 변경내용을 확인합니다.

```python
sejong['총인구수'] = sejong['총인구수'].str.replace(',', '', regex=True).astype(int)
sejong['남자'] = sejong['남자'].str.replace(',', '', regex=True).astype(int)
sejong['여자'] = sejong['여자'].str.replace(',', '', regex=True).astype(int)
sejong
```

### 8. 읍면동별로 세종 인구를 차트로 표시합니다.(인구순 5개만)

```python
import matplotlib.pyplot as plt
plt.rc('font', family='NanumBarunGothic')
```

```python
sejong = sejong.sort_values(by='총인구수', ascending=False)
```

```python
sejong_top5 = sejong.head(5)
```

```python
plt.figure(figsize=(10, 6))
plt.bar(sejong_top5['행정구역'], sejong_top5['총인구수'])
plt.title('세종특별자치시 읍면동별 총인구수 (상위 5개)')
plt.xticks(rotation=45)
plt.xlabel('읍면동')
plt.ylabel('총인구수')
plt.tight_layout()
plt.show()
```

### 9. Choropleth, geojson 등을 이용하여 세종특별자치시의 동별 인구수를 색상별로 지도에 나타내시오

```python
import json
import folium
```

```python
jsonfile = open('HangJeongDong_ver20230101.geojson','r',encoding='utf8')
jsondata = json.load(jsonfile)

jsondata_sejong = {'type':'FeatureCollection'}
jsondata_pick=[]
for item in jsondata['features']:
    if item['properties']['sidonm'] == '세종특별자치시':
        dong = item['properties']['adm_nm'].split()[-1].strip()
        item['id'] = dong
        jsondata_pick.append(item)
jsondata_sejong['features'] = jsondata_pick

with open ('sejong_geojson.json','w')as f:
    json.dump(jsondata_sejong,f)
```

```python
map=folium.Map((36.6108968,127.2864676),zoom_start=10)
jsonfile=open('sejong_geojson.json','r')
jsondata=json.load(jsonfile)
folium.Choropleth(geo_data=jsondata,data=sejong.iloc[2:],key_on='feature.id',bins=5,columns=['동','총인구수'],legend_name='세종 인구수').add_to(map)
map
```
